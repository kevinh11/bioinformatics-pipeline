```python
import os
import dicom2nifti
import pydicom
import highdicom
from nibabel.processing import resample_from_to
import nibabel as nib
import numpy as np
import pandas as pd
import json

base = os.getcwd()
filename_track_df = pd.read_csv("filename_track.csv")

patient_list = df["Subject ID_x"].unique()

# Trackers
last_output_folder = None
last_segment_dir = None
last_affine_report = []


def get_segmentation_location(scan_dir: str) -> str:
    """Find DICOM segmentation file next to a phase directory."""
    parent = os.path.dirname(scan_dir)
    for item in os.listdir(parent):
        if "-Segmentation" in item:
            seg_path = os.path.join(parent, item, "1-1.dcm")
            if os.path.exists(seg_path):
                return seg_path
    print(f"[WARN] No segmentation found in {parent}")
    return None


def get_new_fp(new_fp_dir, old_fp: str) -> str:
    file_name = os.path.basename(old_fp)
    fp_num, fp_desc = file_name.split("-", 1)
    number = str(int(float(fp_num)))
    new_fname_parts = [number, "_".join(fp_desc.lower().split(" "))]
    return os.path.join(new_fp_dir, "_".join(new_fname_parts))


def convert_scans_to_nifti(patient_rows: pd.DataFrame):
    global filename_track_df
    global last_output_folder, last_segment_dir, last_affine_report

    for _, row in patient_rows.iterrows():
        patient_phase_path = os.path.join(base, row["File Location"])
        output_folder = os.path.join(base, "Niftis - Resampling", row["Subject ID_x"])
        os.makedirs(output_folder, exist_ok=True)
        last_output_folder = output_folder

        # --- Convert CT DICOM to NIfTI ---
        new_fp = get_new_fp(output_folder, patient_phase_path)
        filename_dict = dicom2nifti.convert_directory(
            patient_phase_path,
            output_folder,
            compression=True,
            reorient=True
        )
        print(f"[INFO] Converted CT for {row['Subject ID_x']} -> {filename_dict}")

        # Track CT path for later alignment
        ct_ref_path = list(filename_dict.values())[0]
        ct_ref = nib.load(ct_ref_path)

        fp_map_ct = {
            "Subject ID": [row["Subject ID_x"]],
            "old fp": [row["File Location"]],
            "new_fp": [new_fp],
            "modality": ["CT"]
        }


        filename_track_df = pd.concat(
            [filename_track_df, pd.DataFrame.from_dict(fp_map_ct)],
            axis=0
        )


    filename_track_df.to_csv("filename_track.csv", index=False)

import gc
# --- Run ---
for pid in patient_list[:72]:

    if pid == "HCC_085":
        continue

    patient_df_rows = df[df["Subject ID_x"] == pid]
    try:
        convert_scans_to_nifti(patient_df_rows)
    except FileNotFoundError as e:
        print(e)
        continue

    gc.collect()
```

```python
import SimpleITK as sitk
import os

files_dir = os.path.join(os.getcwd(), "Niftis")

def resample_to_isotropic(files_dir, interpolator, patient_id, files_to_skip = []):
    """Resample all NIfTI files for a patient in place to isotropic spacing."""
    # patient_dir = os.path.join(files_dir, patient_id)

    for file in os.listdir(files_dir):
        if not (file.endswith(".nii") or file.endswith(".nii.gz")):
            continue

        file_path = os.path.join(files_dir, file)

        fp_parts = file_path.split(".")

        # new_fp = ".".join(fp_parts[:-2]) + "_resampled.nii.gz"

        if file_path in files_to_skip:
            print("Skipping file:", file_path)
            continue


        print("Processing file:", file_path)

        sitk_img = sitk.ReadImage(file_path)
        original_spacing = sitk_img.GetSpacing()
        original_size = sitk_img.GetSize()
        new_spacing = (1.0, 1.0, 1.0)

        # Compute new image size based on spacing ratios
        new_size = [
            int(round(osz * (ospc / nspc)))
            for osz, ospc, nspc in zip(original_size, original_spacing, new_spacing)
        ]

        # Configure resampling filter
        resample_filter = sitk.ResampleImageFilter()
        resample_filter.SetInterpolator(interpolator)
        resample_filter.SetOutputSpacing(new_spacing)
        resample_filter.SetSize(new_size)
        resample_filter.SetOutputDirection(sitk_img.GetDirection())
        resample_filter.SetOutputOrigin(sitk_img.GetOrigin())
        resample_filter.SetDefaultPixelValue(0)
        resample_filter.SetOutputPixelType(sitk_img.GetPixelID())

        # Execute and overwrite
        resampled_img = resample_filter.Execute(sitk_img)
        sitk.WriteImage(resampled_img, file_path, useCompression=True)

        print(f"  Original spacing: {original_spacing}")
        print(f"  Resampled spacing: {resampled_img.GetSpacing()}")
        print(f"  Overwritten: {file_path}")



patient_list = df["Subject ID_x"].unique()

for pid in patient_list:
    if pid == "resampled":
        continue


    if pid in ["HCC_082", "HCC_085", "HCC_065"]:
      continue

    print(f"Resampling Patient ID: {pid}")

    ct_scan_dir = os.path.join(files_dir, pid)
    seg_dir = os.path.join(files_dir, pid, "segments")

    resample_to_isotropic(ct_scan_dir, sitk.sitkLinear, pid, ["segmentations"])
    resample_to_isotropic(seg_dir, sitk.sitkNearestNeighbor, pid, [])
    # resample_to_isotropic(organ_seg_dir, sitk.sitkNearestNeighbor, pid)

    #NEXT TO CONVERT : HCC_009
    
```
