```python
import pandas as pd
import numpy as np
```

```python
clin_df = pd.read_csv("clinical_df_cleaned.csv").drop(columns=["Unnamed: 0", "TCIA_ID"])

clin_df["empty_fu"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8)) # Adjust the size for readability

sns.heatmap(
    clin_df.corr(),
    annot=True,         # Show the correlation values on the plot
    fmt=".2f",          # Format the annotations to two decimal places
    cmap='coolwarm',    # Color map (coolwarm is standard for correlations: blue for positive, red for negative)
    linewidths=0.5,     # Add lines between cells for clarity
    linecolor='black',
    cbar=True           # Show the color bar
)

# --- 3. Finalize and Display ---
plt.title('Correlation Matrix of All Variables')
plt.show()
```
