```python
import gc
import traceback
import numpy as np
import os
import pandas as pd
import SimpleITK as sitk
from radiomics import featureextractor
from pathlib import Path
import tempfile

# --- PyRadiomics setup ---
config_file = "radiomics_params.yaml"
extractor = featureextractor.RadiomicsFeatureExtractor(config_file)

RADIOMICS_OUTPUT_FILE = "radiomics_features_updated.csv"

hu_lower_bound = -100
hu_upper_bound = 300

patients_wo_pv_alignment = []


def extract_patient_radiomics(data: pd.DataFrame, patient_id: str):
    temp_dir = Path(tempfile.gettempdir())
    patient_row = data[data["Subject ID_x"] == patient_id]

    # --- LOAD CTs (NOW USING sitk.ReadImage, SAME AS SCRIPT 2) ---
    cts_sitk = {}

    for _, row in patient_row.iterrows():
        print(row)
        img_path = f"{row['new_fp_fix']}.nii.gz"
        ct_sitk = sitk.ReadImage(img_path)
        cts_sitk[row["phase"]] = ct_sitk

    if "pv" not in cts_sitk:
        patients_wo_pv_alignment.append(patient_id)
        raise RuntimeError(f"Patient {patient_id} missing PV phase")

    pv_ref = cts_sitk["pv"]

    # --- LOAD SEGMENTATIONS (ALSO USING sitk.ReadImage LIKE SCRIPT 2) ---
    seg_dir = os.path.join(os.path.dirname(patient_row.iloc[0]["new_fp_fix"]), "segments")
    segs_sitk = {}

    for seg_name in ("Liver", "Mass"):
        seg_path = os.path.join(seg_dir, f"Segmentation-{seg_name}.nii.gz")
        seg_sitk = sitk.ReadImage(seg_path)
        seg_sitk = sitk.Cast(seg_sitk, sitk.sitkUInt8)
        segs_sitk[seg_name] = seg_sitk

    all_features = {}

    for phase, ct_sitk in cts_sitk.items():

        # 1. HU clipping (Script 2)
        ct_np = sitk.GetArrayFromImage(ct_sitk).astype(np.float32)
        ct_np = np.clip(ct_np, hu_lower_bound, hu_upper_bound)
        ct_clipped = sitk.GetImageFromArray(ct_np)
        ct_clipped.CopyInformation(ct_sitk)

        # 2. Resample CT → PV geometry (Script 2)
        if phase == "pv":
            ct_aligned = ct_clipped
        else:
            ct_aligned = sitk.Resample(
                ct_clipped,
                pv_ref,
                sitk.Transform(),
                sitk.sitkLinear,
                0,
                ct_clipped.GetPixelID()
            )

        # 3. Segments
        for seg_name, seg_sitk in segs_sitk.items():

            seg_np = sitk.GetArrayFromImage(seg_sitk)
            seg_bin = (seg_np > 0).astype(np.uint8)
            seg_bin_sitk = sitk.GetImageFromArray(seg_bin)
            seg_bin_sitk.CopyInformation(seg_sitk)

            # Resample mask to PV (script 2)
            seg_aligned = sitk.Resample(
                seg_bin_sitk,
                pv_ref,
                sitk.Transform(),
                sitk.sitkNearestNeighbor,
                0,
                sitk.sitkUInt8
            )

            # ---- TEMP FILES (Script 2 requirement) ----
            tmp_ct = temp_dir / f"tmp_ct_{patient_id}_{phase}.nii.gz"
            tmp_mask = temp_dir / f"tmp_mask_{patient_id}_{phase}_{seg_name}.nii.gz"

            sitk.WriteImage(ct_aligned, str(tmp_ct))
            sitk.WriteImage(seg_aligned, str(tmp_mask))

            features = extractor.execute(str(tmp_ct), str(tmp_mask))

            # Cleanup
            if tmp_ct.exists(): tmp_ct.unlink()
            if tmp_mask.exists(): tmp_mask.unlink()

            prefix = f"{phase}_{seg_name}_"
            for k, v in features.items():
                all_features[prefix + k] = v

    all_features["Subject ID_x"] = patient_id
    return all_features



def update_radiomic_csv(batch_dict):
    new_df = pd.DataFrame([batch_dict])
    pid = new_df["Subject ID_x"].iloc[0]

    if os.path.exists(RADIOMICS_OUTPUT_FILE):
        old_df = pd.read_csv(RADIOMICS_OUTPUT_FILE)
        old_df = old_df[old_df["Subject ID_x"].astype(str) != pid]
        merged = pd.concat([old_df, new_df], ignore_index=True)
    else:
        merged = new_df

    merged.to_csv(RADIOMICS_OUTPUT_FILE, index=False)


# ---------------------------------------------------
# MAIN LOOP — unchanged
# ---------------------------------------------------
for pid in os.listdir(os.path.join(os.getcwd(), "Niftis")):
    try:
        patient_feats = extract_patient_radiomics(merged_info_use, pid)
        update_radiomic_csv(patient_feats)
    except Exception as e:
        traceback.print_exc()
        print(e)
        continue
    gc.collect()
    

```
