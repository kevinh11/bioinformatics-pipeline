```python
from statsmodels.stats.inter_rater import fleiss_kappa
import numpy as np
import pandas as pd

# Get ratings and drop rows where all raters are missing
mrecist_ratings = clin_df[["1_mRECIST", "2_mRECIST", "3_mRECIST"]].dropna(how="all", axis=0)

# Convert to binary (1 if > 2, else 0) and count responses
mrecist_binary = (mrecist_ratings > 2).astype(int)
n_categories = 2  # 0=non-responsive, 1=responsive

# Create a count matrix where each row sums to the number of raters (3)
n_subjects = len(mrecist_ratings)
count_matrix = np.zeros((n_subjects, n_categories), dtype=int)

# For each subject, count how many raters chose each category
for i in range(n_subjects):
    counts = mrecist_binary.iloc[i].value_counts().reindex([0,1], fill_value=0)
    count_matrix[i] = counts

# Now compute Fleiss' Kappa
kappa = fleiss_kappa(count_matrix)
print(f"Fleiss' Kappa: {kappa:.3f}")
```

```python
def get_rows_by_num_readers(df : pd.DataFrame, reader_num : int) -> list[int]:
    readers = [1,2,3]
    reader_count = 0
    indices = []
    for idx, _ in enumerate(df):
        # print(df.iloc[idx])
        for r in readers:
            if df.iloc[idx][f"{r}_mRECIST"] != np.nan:
                reader_count += 1

            if reader_count == reader_num:
                indices.append(idx)

    return indices

print(get_rows_by_num_readers(clin_df, 2))
```

```python
def is_binary_float(arr: np.ndarray) -> bool:
    """Check if array only contains 0.0/1.0 (ignoring NaN)."""
    unique_vals = pd.Series(arr).dropna().unique()
    return set(unique_vals).issubset({0.0, 1.0})


def seperate_numerical_ordinal_cols(df: pd.DataFrame):
    numerical = []
    categorical = []
    for col in df:
        if is_binary_float(df[col].values):
            categorical.append(col)   # treat as categorical/boolean
        else:
            numerical.append(col)     # everything else numerical
    return numerical, categorical


numerical ,categorical = seperate_numerical_ordinal_cols(clin_df)

print(numerical)
print(categorical)

```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

def rename_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    Creates shadow features by permuting each column.
    """
    shadow_df = pd.DataFrame(index=df.index)
    for col in df.columns:
        shadow_df[f"shadow_{col}"] = np.random.permutation(df[col].values)
    return shadow_df

feature_importances = pd.DataFrame()

def boruta_fe(xt: pd.DataFrame, yt: pd.Series, n_iter=50):
    global feature_importances

    importances_list = []
    for idx in range(n_iter):
        shadow = rename_features(xt)
        new_xt = pd.concat([xt, shadow], axis=1)

        if feature_importances.empty:
            feature_importances = pd.DataFrame(columns=new_xt.columns)

        rf = RandomForestClassifier(n_estimators=500)
        rf.fit(new_xt, yt)
        importances_list.append(rf.feature_importances_)

    feature_importances = pd.DataFrame(importances_list, columns=new_xt.columns)
    # return average importances across runs
    return feature_importances.mean().sort_values(ascending=False)

# run
avg_importances = boruta_fe(x_train, y_train, n_iter=100)
print(avg_importances)
```

```python


from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import warnings
from sklearn.exceptions import ConvergenceWarning, DataConversionWarning


def execute_rfe_holdout(feature_limit, columns, classifier_tuple, x_train, y_train,
                        x_test, y_test):
    features = columns
    # single holdout split
    # track accuracy per iteration
    acc_history = []
    accuracy_leaderboard = []



    while len(features) >= feature_limit:
        X_train_sub = x_train[features]
        X_test_sub = x_test[features]

        clf = Pipeline([
            classifier_tuple
        ])

        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", category=ConvergenceWarning)
            warnings.filterwarnings("ignore", category=DataConversionWarning)
            clf.fit(X_train_sub, y_train)

        y_pred = clf.predict(X_test_sub)
        acc = accuracy_score(y_test, y_pred)
        acc_history.append(acc)

        # Feature weights from the classifier
        est = clf.named_steps[classifier_tuple[0]]
        weights = np.abs(est.coef_).ravel()
        feature_ranking = sorted(zip(features, weights), key=lambda x: x[1], reverse=True)
        least_useful_feature = feature_ranking[-1][0]

        print(f"Remaining {len(features)} features, accuracy={acc:.4f}, dropping: {least_useful_feature}")
        features.remove(least_useful_feature)

    # plot accuracy progression
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(acc_history)), acc_history, marker="o")
    plt.xlabel("Iteration (features dropped)")
    plt.ylabel("Accuracy on holdout test")
    plt.title("RFE Accuracy Progression")
    plt.grid(True)
    plt.show()

    return acc_history, features


accuracies, final_features = execute_rfe_holdout(
    feature_limit=1,
    columns=list(x_train.columns),
    classifier_tuple=('svc', LinearSVC(penalty='l2', max_iter=5000)),
    x_train=x_train,
    y_train=y_train,
    x_test=x_test,
    y_test = y_test
)
```

```python

from sklearn.metrics import average_precision_score

def execute_forward_selection(columns, x_train, y_train, x_test, y_test, alg, iterations, max_feats):
    accuracy_leaderboard = []
    auc_prc_leaderboard = []
    
    # Store all accuracies and AUC-PRC per iteration for plotting
    all_accuracies = []
    all_auc_prcs = []

    feature_top_n_counts = {k : 0 for k in columns}


    for i in range(iterations):
        feature_set = columns.copy()
        np.random.shuffle(feature_set)

        selected_features = []
        accuracies = []
        auc_prcs = []

        for idx, feat in enumerate(feature_set):
            selected_features.append(feat)
            x_to_process = x_train[selected_features]
            alg.fit(x_to_process, y_train)
            y_pred = alg.predict(x_test[selected_features])
            
            acc = accuracy_score(y_test, y_pred)
            auc = average_precision_score(y_test, y_pred)
            accuracies.append(acc)
            auc_prcs.append(auc)

        all_accuracies.append(accuracies)
        all_auc_prcs.append(auc_prcs)

        best_idx = np.argmax(accuracies)
        best_idx_auc = np.argmax(auc_prcs)
        best_features = selected_features[:min(best_idx+1, max_feats)]

        accuracy_leaderboard.append({
            "iteration": i+1,
            "acc": accuracies[best_idx],
            "features": best_features
        })

        auc_prc_leaderboard.append({
            "iteration": i+1,
            "acc": auc_prcs[best_idx_auc],
            "features": best_features
        })

        for feat in best_features:
            feature_top_n_counts[feat] += 1

    # Plot Accuracy
    plt.figure(figsize=(10, 5))
    for i, acc_list in enumerate(all_accuracies):
        plt.plot(range(1, len(acc_list)+1), acc_list, alpha=0.5, label=f'Iteration {i+1}')
    plt.xlabel("Features Added")
    plt.ylabel("Accuracy")
    plt.title("Forward Feature Selection - Accuracy")
    plt.legend(loc="lower right", fontsize="small")
    plt.show()

    # Plot AUC-PRC
    plt.figure(figsize=(10, 5))
    for i, auc_list in enumerate(all_auc_prcs):
        plt.plot(range(1, len(auc_list)+1), auc_list, alpha=0.5, label=f'Iteration {i+1}')
    plt.xlabel("Features Added")
    plt.ylabel("AUC-PRC")
    plt.title("Forward Feature Selection - Average Precision")
    plt.legend(loc="lower right", fontsize="small")
    plt.show()

    return accuracy_leaderboard, auc_prc_leaderboard, feature_top_n_counts

alg = LinearSVC(penalty='l2', max_iter=5000)
accuracy_leaderboard,auc_prc_leaderboard, feature_counts = execute_forward_selection(list(x_train.columns), x_train, y_train, x_test, y_test, alg, 20, 20)

# print(accuracy_leaderboard)   
```
